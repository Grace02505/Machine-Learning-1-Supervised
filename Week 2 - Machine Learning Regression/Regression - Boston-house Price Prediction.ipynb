```python
for col in ['Alley','MasVnrType','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
            'Electrical','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond',
            'PoolQC','Fence','MiscFeature']:
  train[col]=train[col].fillna('None')
```

Label encode

```python
label_encoder = LabelEncoder()
columns_to_encode=['ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
                   'HeatingQC','KitchenQual','Functional','FireplaceQu','GarageQual','GarageCond','PoolQC','Fence']
categorical_data[columns_to_encode] = categorical_data[columns_to_encode].apply(label_encoder.fit_transform)
categorical_data
```

one hot encoding

```python
cat_one = [x for x in categorical_data if x not in columns_to_encode]

cat_one_encoded = []
for col in cat_one:
  cat_one_encoded += [f"{col[0]}_{cat}" for cat in list(categorical_data[col].unique())]

cat_one_encoded

```

```python
cat_one
for x in cat_one:
  print(x,':',len(train[x].unique()))
```

```python
from sklearn.preprocessing import OneHotEncoder
bee = OneHotEncoder(sparse_output=False)
array = bee.fit_transform(categorical_data[cat_one])
df_bee = pd.DataFrame(array,index = categorical_data.index,columns=cat_one_encoded)
df_bee
```

drop the one hot encodes

```python
enc_df = categorical_data.drop(cat_one,axis=1)

enc_df = pd.concat([enc_df,df_bee],axis=1)
enc_df
```

```python
other_df = train.select_dtypes(exclude='object')

final_df = pd.concat([other_df,enc_df],axis=1)
final_df
```

```python
final_df.isnull().sum() > 0
```

feature selection

```python
X = final_df.drop(labels = ['SalePrice'],axis=1)
X
```

```python
y = final_df['SalePrice']
```

```python
from sklearn.feature_selection import VarianceThreshold
thresh = VarianceThreshold(threshold=0.5)
X_high = thresh.fit_transform(X)
X_high
```

```python
arr = thresh.get_support(indices=True)
select = X.iloc[:,arr].columns.tolist()
select
```

```python
new = final_df[select]
new
```

```python
phoy = new.columns
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
new[phoy] = scaler.fit_transform(new[phoy])
new
```

model the data

```python
X = new
y = final_df['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

model = LinearRegression()
model.fit(X_train, y_train)
expected_y = y_test
predicted_y = model.predict(X_test)
```

```python
print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_absolute_error(expected_y, predicted_y))
```
